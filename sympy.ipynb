```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sympy.symbols("x")
```

```python
f = sympy.sin(x)/x
```

```python
F = sympy.lambdify(x, f)
```

```python
t, dt = np.linspace(1e-3, 15, 1_000, retstep=True)
```

```python
y = F(t)
```

```python
integral_y = np.cumsum(y)*dt
```

```python
arc_length_y = np.cumsum(np.sqrt(1 + np.diff(y/dt)**2))*dt
```

```python
%matplotlib widget
fig, ax = plt.subplots()

ax.plot(t, y, color="blue")
ax.plot(t, integral_y, color="red")
# ax.plot(t[:-1], arc_length_y, color="black")
ax.axis("equal")
ax.legend(["function", "definite integral", "arc length"])
plt.show()
```
